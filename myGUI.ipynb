```python
import PySimpleGUI as sg
import os.path
from PIL import Image
import cv2
import time

# Normally tkinter only wants PNG and GIF files.  
# This program uses PIL to convert files such as jpg files into a PNG format so that tkinter can use it.


# ========================================================================================================#


sg.theme('DefaultNoMoreNagging')

left_col = [[sg.Image(r'C:\Users\User\Desktop\Py_GUI\logo1.png')],
            [sg.Button('Take a Picture')],
            [sg.Button('Recognize the Object')],
            [sg.Button('Grasp the Object')],
            [sg.Text('\n')],
            [sg.Button('Clear'), sg.Button('Exit')],
            #[sg.Text('\n Dr. Henry Chu\'s Lab\n ME@PolyU x HUAWEI',justification='center')]]
            [sg.Text('\n\n ME@PolyU x HUAWEI',justification='center')]]

r1_col = [[sg.Text('Captured Image')],
          [sg.Image(r'C:\Users\User\Desktop\Py_GUI\black.png', key='_R1_', size=(400, 300))]]
r2_col = [[sg.Text('Object Recognition')],
          [sg.Image(r'C:\Users\User\Desktop\Py_GUI\black.png', key='_R2_', size=(400, 300))]]
r3_col = [[sg.Text('Real-Time Camera View')],
          [sg.Image(r'C:\Users\User\Desktop\Py_GUI\black.png', key='_R3_', size=(400, 300))]]


layout = [[sg.Column(left_col, element_justification='c'),
           sg.VSeperator(),
           sg.Column(r1_col, element_justification='c'),
           sg.Column(r2_col, element_justification='c'),
           sg.Column(r3_col, element_justification='c')]]

window = sg.Window('demo', layout, resizable=False)

camera_Width  = 400 # 320 # 480 # 640 # 1024 # 1280
camera_Heigth = 300 # 240 # 320 # 480 # 780  # 960
frameSize = (camera_Width, camera_Heigth)
camera_on = 0;
cap = cv2.VideoCapture(0)


while True:
    start_time = time.time()
    [event, values] = window.read(timeout = 10)
    ret, frameOrig = cap.read()
    frame = cv2.resize(frameOrig, frameSize)
    imgbytes = cv2.imencode('.png', frame)[1].tobytes()
    window['_R3_'].update(data=imgbytes)
    
    
    if event in (None, 'Exit'):
        break
    if event == 'Clear':
         window['_R1_'].update(r'C:\Users\User\Desktop\Py_GUI\black.png', size=(400, 300))
         window['_R2_'].update(r'C:\Users\User\Desktop\Py_GUI\black.png', size=(400, 300))
        

    if event == 'Take a Picture':
        #==========================#
        # Take a picture using the camera.
        
        for i in range(1):
            return_value, image = cap.read()
            cv2.imwrite('opencv'+str(i)+'.png', image)
     #   del(camera)
        
        #==========================#
        
        MAX_SIZE = (400, 300)
        image = Image.open(r'C:\Users\User\Desktop\Py_GUI\opencv0.png')
        image.thumbnail(MAX_SIZE)
        # creating thumbnail
        image.save('capture_img_resize.png')
        window['_R1_'].update(r'C:\Users\User\Desktop\Py_GUI\capture_img_resize.png')
        
    if event == 'Recognize the Object':
        #==========================#
        
        # Code for Image Processing.
        # Output a processed image.
        
        #==========================#
        
        MAX_SIZE = (400, 300)
        image = Image.open(r'C:\Users\User\Desktop\Py_GUI\favicon.png')
        image.thumbnail(MAX_SIZE)
        # creating thumbnail
        image.save('reg_obj_resize.png')
        window['_R2_'].update(r'C:\Users\User\Desktop\Py_GUI\reg_obj_resize.png')        
        
    if event == 'Grasp the Object':
        #==========================#
        
        # Code for Image Processing.
        # Output a processed image.
        
        #==========================#
        break
                 
# --------------------------------- Close & Exit ---------------------------------  

cap.release()
cv2.destroyAllWindows()
window.close()
```
